```python
#################### Full Project code ############################
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv("students_marksss.csv")

# 2. Add total & average
df["Total_Marks"] = df["Maths"] + df["Science"] + df["English"]
df["Average_Marks"] = df["Total_Marks"] / 3

print(df)

# 3. Bar chart: Average per student
plt.figure()
plt.bar(df["Student"], df["Average_Marks"])
plt.title("Average Marks per Student")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()

# 4. Bar chart: Subject-wise averages
subject_avgs = df[["Maths", "Science", "English"]].mean()

plt.figure()
plt.bar(subject_avgs.index, subject_avgs.values)
plt.title("Average Marks per Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

# 5. Line + Pie: One student (H)
student_h = df[df["Student"] == "H"][["Maths", "Science", "English"]].iloc[0]
subjects = ["Maths", "Science", "English"]
marks = [student_h["Maths"], student_h["Science"], student_h["English"]]

plt.figure()
plt.plot(subjects, marks, marker='o')
plt.title("Student H - Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

plt.figure()
plt.pie(marks, labels=subjects, autopct="%1.1f%%")
plt.title("Student H - Subject-wise Percentage")
plt.show()

# 6. Scatter: Attendance vs Average
plt.figure()
plt.scatter(df["Attendance"], df["Average_Marks"])
plt.title("Attendance vs Average Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Average Marks")
plt.show()

# 7. Top 3 students
top_3 = df.sort_values(by="Average_Marks", ascending=False).head(3)
print("Top 3 Students:")
print(top_3[["Student", "Average_Marks"]])
```

```python
# Add a new column: Result (Pass/Fail, pass if Average ≥ 60).
# After you calculate Average_Marks
df["Result"] = df["Average_Marks"].apply(lambda x: "Pass" if x >= 60 else "Fail")

print(df[["Student", "Average_Marks", "Result"]])
```

```python
# Draw a bar chart showing only students who failed.
failed_students = df[df["Result"] == "Fail"]

print(failed_students[["Student", "Average_Marks", "Result"]])

import matplotlib.pyplot as plt

plt.figure()

plt.bar(failed_students["Student"], failed_students["Average_Marks"])

plt.title("Failed Students - Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")

plt.show()
```

```python
##################################################### Nothing ####################################3
import matplotlib.pyplot as plt

days = [1,2,3,4]
marks1 = [70,75,80,85]
marks2 = [65,72,78,82]
marks3 = [68,70,74,78]

plt.figure(figsize=(12,5))

plt.plot(days, marks1, color='deeppink', marker='o',
         linewidth=3, markersize=8, label='Student A')
plt.fill_between(days, marks1, color='deeppink', alpha=0.2)

plt.plot(days, marks2, color='gold', marker='s',
         linewidth=3, markersize=8, label='Student B')
plt.fill_between(days, marks2, color='gold', alpha=0.2)

plt.plot(days, marks3, color='royalblue', marker='D',
         linewidth=3, markersize=8, label='Student C')
plt.fill_between(days, marks3, color='royalblue', alpha=0.2)

plt.title("Student Performance", fontsize=18)
plt.xlabel("Days", fontsize=13)
plt.ylabel("Marks", fontsize=13)

plt.grid(linestyle='--', alpha=0.5)
plt.legend(shadow=True)

for x, y in zip(days, marks1):
    plt.text(x, y+0.5, y)

plt.show()
```
