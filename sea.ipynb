```python
########################### KDE = Kernel Density Estimate #######################################

# It's basically a smooth curve that shows the probability distribution of data.

# It’s used to understand:

# ✔ Shape of the data
# ✔ Peak values
# ✔ Skewness
# ✔ Spread
# ✔ Multi-modal distribution

# KDE is like a smoother, elegant version of a histogram.
```

```python
# Why is Heatmap Important?

# Heatmaps are used to visualize correlation between numeric columns.
# Correlation helps you understand:
# Which feature affects the target
# Multicollinearity (two columns too similar)
# Which columns are useless
# Which features are important for ML
# Correlation values range from:
# +1 → Strong positive 
# -1 → Strong negative 
# 0 → No relationship
```

```python
df.head()
```

```python
# Step 1: Create Correlation Matrix
corr = df.corr(numeric_only=True)
corr
```

```python
# Step 2: Plot Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()

# Explanation:
# annot=True → shows numbers
# cmap="coolwarm" → color theme
```

```python
# heatmap without annotation
sns.heatmap(corr)
plt.show()
```

```python
# Heatmap With Figure Size
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
import numpy as np
```

```python
# Mask Upper Triangle (Cleaner Look)
# Often we hide half the matrix because it's symmetric.
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, cmap="coolwarm", mask=mask)
plt.show()
```

```python
# How ML Uses Heatmap
# High correlation with target → good feature
# High correlation between two features → drop one
# Low correlation → not useful

# Heatmap is a core step in EDA for ML models.
```

```python
# What is a Pairplot? = EDA
# A pairplot creates:
# Histogram for every variable
# Scatter plot between every pair of numeric variables
# KDE
# Optionally different colors (hue) for categories
# tells about heatmap correlation
# It's like pressing one button and getting full relationship analysis of your dataset.
```

```python
# Why Pairplot is Important?

# ✔ Shows correlation visually
# ✔ Shows cluster patterns
# ✔ Shows linear/non-linear relationships
# ✔ Shows outliers
# ✔ Shows distribution
# ✔ Helps identify feature importance for ML
# This is used heavily in ML EDA (Exploratory Data Analysis).
```

```python
# basic plot
sns.pairplot(df)
plt.show()
# This automatically selects numeric columns.
```

```python
# with hue
sns.pairplot(data=df, hue='sex')
plt.show()
```

```python
# select specific columns
sns.pairplot(df[['total_bill', 'tip', 'size']])
plt.show()
```

```python
# styling pairplot
sns.pairplot(df, hue='smoker', diag_kind='kde')
plt.show()
# diag_kind='kde' changes diagonal from histogram → smooth KDE.
```

```python
# with marker
sns.pairplot(df, hue='sex', markers=['o','s'])
plt.show()
```

```python
# Big Plot → Add Figure Size (Optional Trick)
# Pairplots don’t have a direct figsize,
# but you can change default figure scale:

sns.set(rc={'figure.figsize':(8,6)})
sns.pairplot(df)
```
